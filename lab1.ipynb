## Necessary librares

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Activation functions

### ReLU

```python
def rectified_linear_unit(x):
    return np.maximum(0, x)

def rectified_linear_unit_derivative(x):
    return np.where(x <= 0, 0, 1)
```

```python
x_values = np.linspace(-15, 15, 100)


relu_values = rectified_linear_unit(x_values)
relu_derivative_values = rectified_linear_unit_derivative(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, relu_values, label='ReLU(x)', color='purple', linewidth=2)
plt.plot(x_values, relu_derivative_values, label="ReLU'(x)", color='orange', linewidth=2)
plt.title('Функція ReLU та її похідна')
plt.xlabel('x')
plt.ylabel('ReLU(x) / ReLU\'(x)')
plt.grid(True)
plt.legend()

plt.show()
```

### ReLU6

```python
def rectified_linear_unit_6(x):
    return np.minimum(np.maximum(0, x), 6)

def rectified_linear_unit_derivative_6(x):
    return np.where((x <= 0) | (x >= 6), 0, 1)
```

```python
x_values = np.linspace(-15, 15, 100)

relu6_values = rectified_linear_unit_6(x_values)
relu6_derivative_values = rectified_linear_unit_derivative_6(x_values)

plt.figure(figsize=(10, 6))
plt.plot(x_values, relu6_values, label='ReLU6(x)', color='purple', linewidth=2)
plt.plot(x_values, relu6_derivative_values, label="ReLU6'(x)", color='orange',  linewidth=2)
plt.title('Функція ReLU6 та її похідна')
plt.xlabel('x')
plt.ylabel('ReLU6(x) / ReLU6\'(x)')
plt.grid(True)
plt.legend()

plt.show()
```

## ELU

```python
def elu(x, alpha=1.0):
    return np.where(x >= 0, x, alpha * (np.exp(x) - 1))

def elu_derivative(x, alpha=1.0):
    return np.where(x >= 0, 1, alpha * np.exp(x))
```

```python
x_values = np.linspace(-15, 15, 100)
alpha = 1.0

elu_values = elu(x_values, alpha)
elu_derivative_values = elu_derivative(x_values, alpha)

plt.figure(figsize=(10, 6))
plt.plot(x_values, elu_values, label='ELU(x)', color='purple', linewidth=2)
plt.plot(x_values, elu_derivative_values, label="ELU'(x)", color='orange', linewidth=2)
plt.title('Функція ELU та її похідна')
plt.xlabel('x')
plt.ylabel('ELU(x) / ELU\'(x)')
plt.grid(True)
plt.legend()

plt.show()
```
