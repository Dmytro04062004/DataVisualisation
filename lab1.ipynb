```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x <= 0, 0, 1)

x_values = np.linspace(-5, 5, 100)

relu_values = relu(x_values)
relu_derivative_values = relu_derivative(x_values)

plt.figure(figsize=(10, 5))
plt.plot(x_values, relu_values, label='ReLU(x)', color='purple', linewidth=2)
plt.plot(x_values, relu_derivative_values, label="ReLU'(x)", color='orange', linestyle='--', linewidth=2)
plt.title('Функція ReLU та її похідна')
plt.xlabel('x')
plt.ylabel('ReLU(x) / ReLU\'(x)')
plt.grid(True)
plt.legend()

plt.show() вв



```
